```python
dataset_copy=df_swiggy.copy(deep=True)
```

```python
dataset_copy
```

```python
dataset_copy['type']
```

```python
dataset_copy['Restaurants Type']
```

```python
dataset_copy.rename(columns = {'type':'Restaurants Type'}, inplace = True)
```

```python
def replace(data):
    if data =='F':
        return data.replace("F","Fine")
    elif data =='D':
        return data.replace("D","Dining")
    elif data =='C':
        return data.replace("C","Continental")
    else:
        return data.replace("P","Premium")
```

```python
Type = dataset_copy['Restaurants Type'].apply(lambda x: replace(x))
```

```python
dataset_copy['Restaurants Type']=Type
```

```python
dataset_copy.head(5)
```

```python
cuisines = dataset_copy['cuisines'].apply(lambda x: x.replace(" ",","))
dataset_copy['cuisines'] = cuisines
dataset_copy.head(5)
```

```python
dataset_copy.rename(columns = {'veg':'Vegetarian'}, inplace = True)
```

```python
dataset_copy.head(5)
```

```python
dataset_copy.drop(columns=['City'], inplace=True)
```

```python
dataset_copy.head(5)
```

```python
dataset_copy['avgRating'].isna().sum()
```

```python
avgRating = dataset_copy['avgRating'].apply(lambda s : s.replace('--','0.0'))
dataset_copy['avgRating'] = avgRating
dataset_copy.head()
```

```python
dataset_copy['costForTwoStrings'] = dataset_copy['costForTwoStrings'].astype('int64')
```

```python
costForTwoStrings = dataset_copy['costForTwoStrings'].apply(lambda x: x.replace(x[-8:],'')[1:])
dataset_copy['costForTwoStrings'] = costForTwoStrings
dataset_copy.head(5)
```

```python
totalRatingsString = dataset_copy['totalRatingsString'].apply(lambda s : s.replace(s[-8:],''))
dataset_copy['totalRatingsString'] = totalRatingsString
dataset_copy.head(2)
```

```python
dataset_copy.drop(columns=['totalRatingsString'], inplace=True)
```

```python
dataset_copy.head(2)
```

```python
dataset_copy = dataset_copy[dataset_copy['costForTwoStrings']>100]
dataset_copy
```

```python
dataset_copy.info()
```

```python
dataset_copy.drop(columns=['address'], inplace=True)
```

```python
dataset_copy.head(2)
```

```python
dataset_copy.dropna(subset = ['unserviceable','Vegetarian','locality'],axis=0,inplace=True)
```

```python
dataset_copy.info()
```

```python
dataset_copy = dataset_copy.drop_duplicates()
```

```python
dataset_copy.isna().sum()
```

```python
dataset_copy.shape
```

```python
dataset_copy.to_csv('Swiggy_Final_Data.csv',index=False)
```
